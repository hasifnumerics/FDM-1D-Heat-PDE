```python
import numpy as np
import matplotlib.pyplot as plt

L=20 #length of the object to be 10 units
Nx=120 #Size of x, as in the x coordinate system is split into 128 data points by python
Nt=100 #Size of t, as in after how much time you want to know about the temperature of the object

x=np.linspace(-L/2,L/2,Nx) #Defines x's territory to be from -L/2 to L/2 covering the whole length of the object. The whole space is covered by Nx data points)
dx= x[1]-x[0] #discrete differences
k=1.0
dt=0.1*dx**2/k #alpha= 0.1
u=np.exp(-x**2) #Initial distribution of heat as a function of space

boundary="Dirichlet"
plt.plot(x,u,"b") #initial temperature

for i in range(Nt): #explained in paper-essentially we create the for loop here
    u[1:Nx-1]= u[1:Nx-1]+dt*k*(u[2:Nx]-2*u[1:Nx-1]+u[0:Nx-2])/dx**2
    if boundary=="Dirichlet":
        u[0]=0
        u[Nx-1]=0
        
plt.plot(x,u,"r") #represents the temperature after t time steps
plt.show()
```

```python
#if the initial heat distribution is different
import numpy as np
import matplotlib.pyplot as plt

L=20 #length of the object to be 10 units
Nx=120 #Size of x, as in the x coordinate system is split into 128 data points by python
Nt=1000 

x=np.linspace(-L/2,L/2,Nx) #Defines x's territory to be from -L/2 to L/2 covering the whole length of the object. The whole space is covered by Nx data points)
dx= x[1]-x[0] 
k=1.0
dt=0.1*dx**2/k # a
u=np.exp(-4*x**2) #Initial distribution of heat as a function of space

boundary="Dirichlet"
plt.plot(x,u,"b")

for i in range(Nt):
    u[1:Nx-1]= u[1:Nx-1]+dt*k*(u[2:Nx]-2*u[1:Nx-1]+u[0:Nx-2])/dx**2
    if boundary=="Dirichlet":
        u[0]=0
        u[Nx-1]=0
        
plt.plot(x,u,"y")
plt.show()

#Talked about in the paper- notice how a more spiked initial heat distribution leads to a faster spreading of the heat. This confirms our derivation intuition mentioned in the paper as well.
```
